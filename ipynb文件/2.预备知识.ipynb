### 2.1.5 è¿ç®—å†…å­˜å¼€é”€

å‰é¢è¯´äº†ï¼Œç´¢å¼•æ“ä½œæ˜¯ä¸ä¼šå¼€è¾Ÿæ–°å†…å­˜çš„ï¼Œè€Œåƒ` y = x + y` è¿™æ ·çš„è¿ç®—æ˜¯ä¼šæ–°å¼€å†…å­˜çš„ï¼Œç„¶åå°†yæŒ‡å‘æ–°å†…å­˜ã€‚

ä¸ºäº†æ¼”ç¤ºè¿™ä¸€ç‚¹ï¼Œæˆ‘ä»¬å¯ä»¥ä½¿ç”¨Pythonè‡ªå¸¦çš„ `id` å‡½æ•°ï¼šå¦‚æœä¸¤ä¸ªå®ä¾‹çš„IDä¸€è‡´ï¼Œé‚£ä¹ˆå®ƒä»¬æ‰€å¯¹åº”çš„å†…å­˜åœ°å€ç›¸åŒï¼›åä¹‹åˆ™ä¸åŒã€‚

````python
x = torch.tensor([1, 2])
y = torch.tensor([3, 4])
id_before = id(y)
y = y + x
print(id(y) == id_before) # False 
````

å¦‚æœæƒ³æŒ‡å®šç»“æœåˆ°åŸæ¥çš„ `y` çš„å†…å­˜ï¼Œæˆ‘ä»¬å¯ä»¥ä½¿ç”¨å‰é¢ä»‹ç»çš„ç´¢å¼•æ¥è¿›è¡Œæ›¿æ¢æ“ä½œã€‚åœ¨ä¸‹é¢çš„ä¾‹å­ä¸­ï¼Œæˆ‘ä»¬æŠŠ` x + y `çš„ç»“æœé€šè¿‡ `[:]` å†™è¿› `y` å¯¹åº”çš„å†…å­˜ä¸­ã€‚

````python
x = torch.tensor([1, 2])
y = torch.tensor([3, 4])
id_before = id(y)
y[:] = y + x
print(id(y) == id_before) # True
````

æˆ‘ä»¬è¿˜å¯ä»¥ä½¿ç”¨è¿ç®—ç¬¦å…¨åå‡½æ•°ä¸­çš„ `out` å‚æ•°æˆ–è€…è‡ªåŠ è¿ç®—ç¬¦ `+=` (ä¹Ÿå³ `add_()` )è¾¾åˆ°ä¸Šè¿°æ•ˆæœï¼Œä¾‹å¦‚ `torch.add(x, y, out=y) `å’Œ `y += x(y.add_(x))`ã€‚

````python
x = torch.tensor([1, 2])
y = torch.tensor([3, 4])
id_before = id(y)
torch.add(x, y, out=y) # y += x, y.add_(x)
print(id(y) == id_before) # True
````

âœ¨ æ³¨ï¼šè™½ç„¶ `view` è¿”å›çš„ `Tensor` ä¸æº `Tensor` æ˜¯å…±äº«dataçš„ï¼Œä½†æ˜¯ä¾ç„¶æ˜¯ä¸€ä¸ªæ–°çš„ `Tensor`ï¼ˆå› ä¸º `Tensor` é™¤äº†åŒ…å«dataå¤–è¿˜æœ‰ä¸€äº›å…¶ä»–å±æ€§ï¼‰ï¼ŒäºŒè€…idï¼ˆå†…å­˜åœ°å€ï¼‰å¹¶ä¸ä¸€è‡´ã€‚
